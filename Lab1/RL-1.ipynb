```python
import gym 
import numpy as np
```

```python
env = gym.make('FrozenLake-v1', is_slippery=False)
```

```python
epsilon = 1.0
epsilon_decay = 0.995
epsilon_min = 0.01
total_episodes = 1000
max_steps = 100
alpha = 0.85
gamma = 0.95
```

```python
Q = np.zeros((env.observation_space.n, env.action_space.n))
```

```python
def choose_action(state):
    if np.random.uniform(0, 1) < epsilon:
        return env.action_space.sample() # explore randomly
    else:
        return np.argmax(Q[state, :])    # exploit the best known action
    
```

```python
def update(state, state2, reward, action, done):
    predict = Q[state, action]
    
    if done:
        target = reward
    else:
        target = reward + gamma * np.max(Q[state2, :])
    
    Q[state, action] = Q[state, action] + alpha * (target - predict)
```

```python
for episode in range(total_episodes):
    state = env.reset()
    action = choose_action(state)
    
    for t in range(max_steps):
        next_state, reward, done, _ = env.step(action)
        update(state, next_state, reward, action, done)
        state = next_state
        action = choose_action(state)
        
        if done:
            break
            
        epsilon = max(epsilon_min, epsilon * epsilon_decay)
```

```python
total_reward = 0

for _ in range(100):
    state = env.reset()
    while True:
        action = np.argmax(Q[state, :])
        next_state, reward, done, _ = env.step(action)
        total_reward += reward
        state = next_state
        
        if done: break
            
print("Average Reward: ", total_reward / 100)
```
